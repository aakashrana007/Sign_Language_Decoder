```python
import cv2
import mediapipe as mp



# Connect to webcam
cap = cv2.VideoCapture(0)  # Use 0 or other index depending on your webcam

# Initialize holistic model
with mp.solutions.holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5, model_complexity=1) as holistic:
    
    while cap.isOpened():
        ret, frame = cap.read()

        if not ret:
            print("Failed to receive frame from webcam. Exiting...")
            break

        # Recolor the frame from BGR to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False  # Set flag to False for performance boost

        # Make detections
        results = holistic.process(image)

        # Recolor back to BGR for rendering
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Draw face landmarks
        mp.solutions.drawing_utils.draw_landmarks(image, results.face_landmarks, mp.solutions.face_mesh.FACEMESH_CONTOURS)

        # Draw right hand landmarks
        mp.solutions.drawing_utils.draw_landmarks(image, results.right_hand_landmarks, mp.solutions.hands.HAND_CONNECTIONS)

        # Draw left hand landmarks
        mp.solutions.drawing_utils.draw_landmarks(image, results.left_hand_landmarks, mp.solutions.hands.HAND_CONNECTIONS)

        # Draw pose landmarks
        mp.solutions.drawing_utils.draw_landmarks(image, results.pose_landmarks, mp.solutions.pose.POSE_CONNECTIONS)

        # Show the annotated frame
        cv2.imshow('Holistic Model Detection', image)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp

# Initialize the mediapipe drawing and holistic solutions
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic

# Connect to webcam
cap = cv2.VideoCapture(0)

# Initiate holistic model
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    
    while cap.isOpened():
        ret, frame = cap.read()
        
        if not ret:
            print("Failed to receive frame from webcam. Exiting...")
            break
        
        # Recolor Feed
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        # Make Detections
        results = holistic.process(image)
        
        # Recolor image back to BGR for rendering
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # 1. Draw face landmarks
        mp_drawing.draw_landmarks(image, results.face_landmarks, mp.solutions.face_mesh.FACEMESH_CONTOURS, 
                                  mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1),
                                  mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1))
        
        # 2. Right hand
        mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                  mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4),
                                  mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2))

        # 3. Left Hand
        mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                  mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4),
                                  mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2))

        # 4. Pose Detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS, 
                                  mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4),
                                  mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))
                        
        cv2.imshow('Raw Webcam Feed', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
class_name= "Wakanda"

import cv2
import mediapipe as mp

# Initialize the mediapipe drawing and holistic solutions
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic

# Connect to webcam
cap = cv2.VideoCapture(0)

# Initiate holistic model
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    
    while cap.isOpened():
        ret, frame = cap.read()
        
        if not ret:
            print("Failed to receive frame from webcam. Exiting...")
            break
        
        # Recolor Feed
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        # Make Detections
        results = holistic.process(image)
        
        # Recolor image back to BGR for rendering
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # 1. Draw face landmarks
        mp_drawing.draw_landmarks(image, results.face_landmarks, mp.solutions.face_mesh.FACEMESH_CONTOURS, 
                                  mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1),
                                  mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1))
        
        # 2. Right hand
        mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                  mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4),
                                  mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2))

        # 3. Left Hand
        mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                  mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4),
                                  mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2))

        # 4. Pose Detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS, 
                                  mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4),
                                  mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))

        #Export coordinates into the csv file
        #converting results into array form suitable
        try:
            #extract pose landmark
            pose=results.pose_landmarks.landmark 
            pose_row = list(np.array([[landmark.x,landmark.y,landmark.z,landmark.visibility] for landmark in pose]).flatten())
            #pose is storing all the pose landmarks
            #next we are looping through landmarks in pose and storing each in array for separate values 
            #flatten- extract individual arrays and convert to big single array

            #same thing for face
            #extract face landmarks
            face=results.pose_landmarks.landmark 
            face_row = list(np.array([[landmark.x,landmark.y,landmark.z,landmark.visibility] for landmark in face]).flatten())

            #concate rows
            row=pose_row + face_row
            #append class name into the row
            row.insert(0, class_name)

            #export csv
            #'a' mode is append mode
            with open('coords.csv', mode='a', newline='') as f:
                csv_writer = csv.writer(f, delimiter=',',quotechar='"',quoting=csv.QUOTE_MINIMAL)
                csv_writer.writerow(row)

        except:    
            pass

        cv2.imshow('Raw Webcam Feed', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

## 3.2 Train Machine Learning Classification Model

```python
#for ML pipeline
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.impute import SimpleImputer
#for ML algorithms; we'll use 4 algorithms for better result
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
# Define pipelines with the new imputation strategy
pipelines = {
    'lr': make_pipeline(SimpleImputer(strategy='constant', fill_value=0),
                        StandardScaler(), LogisticRegression()),
    'rc': make_pipeline(SimpleImputer(strategy='constant', fill_value=0),
                        StandardScaler(), RidgeClassifier()),
    'rf': make_pipeline(SimpleImputer(strategy='constant', fill_value=0),
                        StandardScaler(), RandomForestClassifier()),
    'gb': make_pipeline(SimpleImputer(strategy='constant', fill_value=0),
                        StandardScaler(), GradientBoostingClassifier()),
}
```

```python
fit_models = {}  # Create an empty dictionary to store fitted models

for algo, pipeline in pipelines.items():
    model = pipeline.fit(X_train, y_train)  # Fit each pipeline to the training data
    fit_models[algo] = model  # Store the fitted model in the dictionary with the algorithm name as the key
```

```python
import cv2
import mediapipe as mp
import numpy as np
import pandas as pd

# Initialize the mediapipe drawing and holistic solutions
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic

# Connect to webcam
cap = cv2.VideoCapture(0)

# Initiate holistic model
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    
    while cap.isOpened():
        ret, frame = cap.read()
        
        if not ret:
            print("Failed to receive frame from webcam. Exiting...")
            break
        
        # Recolor Feed
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        # Make Detections
        results = holistic.process(image)
        
        # Recolor image back to BGR for rendering
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # 1. Draw face landmarks
        mp_drawing.draw_landmarks(image, results.face_landmarks, mp.solutions.face_mesh.FACEMESH_CONTOURS, 
                                  mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1),
                                  mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1))
        
        # 2. Right hand
        mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                  mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4),
                                  mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2))

        # 3. Left Hand
        mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                  mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4),
                                  mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2))

        # 4. Pose Detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS, 
                                  mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4),
                                  mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))

        #Export coordinates into the csv file
        #converting results into array form suitable
        try:
            #extract pose landmark
            pose=results.pose_landmarks.landmark 
            pose_row = list(np.array([[landmark.x,landmark.y,landmark.z,landmark.visibility] for landmark in pose]).flatten())
            #pose is storing all the pose landmarks
            #next we are looping through landmarks in pose and storing each in array for separate values 
            #flatten- extract individual arrays and convert to big single array

            #same thing for face
            #extract face landmarks
            face=results.pose_landmarks.landmark 
            face_row = list(np.array([[landmark.x,landmark.y,landmark.z,landmark.visibility] for landmark in face]).flatten())

            #concate rows
            row=pose_row + face_row


            '''
            #append class name into the row
            row.insert(0, class_name)

            #export csv
            #'a' mode is append mode
            with open('coords.csv', mode='a', newline='') as f:
                csv_writer = csv.writer(f, delimiter=',',quotechar='"',quoting=csv.QUOTE_MINIMAL)
                csv_writer.writerow(row)
            '''

            #Make Detections using our trained model
            X = pd.DataFrame([row])
            sign_language_class = model.predict(X)[0]
            sign_language_prob = model.predict_proba(X)[0]
            print(sign_language_class, sign_language_prob)

            #Grab Ear coordinates
            coords = tuple(np.multiply(np.array((results.pose_landmarks.landmark[mp_holistic.PoseLandmark.LEFT_EAR].x,results.pose_landmarks.landmark[mp_holistic.PoseLandmark.LEFT_EAR].y)), [640,480]).astype(int))
            
            #render our prediction
            #draw rectangle
            cv2.rectangle(image, 
                          (coords[0], coords[1]+5), # x and y+5 for thicker y coord
                          (coords[0]+ len(sign_language_class)*20, coords[1]-30), #for dynamic rectangle
                          (245, 117, 16), # BGR color
                          -1) #if we want rectangle field color
            
            #similarly for text
            cv2.putText(image, sign_language_class, coords, cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)

        except:    
            pass

        cv2.imshow('Raw Webcam Feed', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

# Lets debug these errors

> For SimpleImputer Error:

When you fit the SimpleImputer and other preprocessing steps, the feature names must be consistent. Make sure the DataFrame you use for prediction has the same columns as the training data.
For this- 

            columns = [f'{axis}{i}' for i in range(1, len(row) // 4 + 1) for axis in ['x', 'y', 'z', 'v']]

            X = pd.DataFrame([row], columns=columns)
            

Purpose: To ensure that the DataFrame used for prediction has the same column names as those used during training.
Explanation: This snippet dynamically generates column names based on the number of features in the row array. Each feature corresponds to the x, y, z, and visibility coordinates of each landmark. The columns are named x1, y1, z1, v1, x2, y2, ..., ensuring consistency with the training data.
Reason: The warning indicated that the SimpleImputer was fitted with feature names, but the data passed during prediction did not have these names. This mismatch can cause issues during the imputation and scaling steps.

> For SymbolDatabase error:

This warning indicates that a deprecated method (GetPrototype) is being used. If this is coming from a third-party library and you can't update the method directly, you can suppress the warning:
For this- 

            import warnings
        
            warnings.filterwarnings("ignore", category=UserWarning, message='SymbolDatabase.GetPrototype() is deprecated')


```python
#suppress warning
import warnings
warnings.filterwarnings("ignore", category=UserWarning, message=" UserWarning: SymbolDatabase.GetPrototype() is deprecated. Please use message_factory.GetMessageClass() instead. SymbolDatabase.GetPrototype() will be removed soon.")

import pandas as pd
import numpy as np
import cv2
import mediapipe as mp
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import pickle

# Initialize the mediapipe drawing and holistic solutions
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic

# Load the trained model
with open('sign_language.pkl', 'rb') as f:
    model = pickle.load(f)

# Connect to webcam
cap = cv2.VideoCapture(0)

# Initiate holistic model
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        ret, frame = cap.read()

        if not ret:
            print("Failed to receive frame from webcam. Exiting...")
            break

        # Recolor Feed
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        # Make Detections
        results = holistic.process(image)

        # Recolor image back to BGR for rendering
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Draw landmarks
        mp_drawing.draw_landmarks(image, results.face_landmarks, mp.solutions.face_mesh.FACEMESH_CONTOURS,
                                  mp_drawing.DrawingSpec(color=(80, 110, 10), thickness=1, circle_radius=1),
                                  mp_drawing.DrawingSpec(color=(80, 256, 121), thickness=1, circle_radius=1))
        mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(80, 22, 10), thickness=2, circle_radius=4),
                                  mp_drawing.DrawingSpec(color=(80, 44, 121), thickness=2, circle_radius=2))
        mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                                  mp_drawing.DrawingSpec(color=(121, 44, 250), thickness=2, circle_radius=2))
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=4),
                                  mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2))

        try:
            # Extract pose landmarks
            pose = results.pose_landmarks.landmark
            pose_row = list(np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in pose]).flatten())

            # Extract face landmarks
            face = results.face_landmarks.landmark
            face_row = list(np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in face]).flatten())

            # Concatenate rows
            row = pose_row + face_row

            # Ensure column names match those used during training
            columns = [f'{axis}{i}' for i in range(1, len(row) // 4 + 1) for axis in ['x', 'y', 'z', 'v']]
            X = pd.DataFrame([row], columns=columns)
            
            # Make predictions
            sign_language_class = model.predict(X)[0]
            sign_language_prob = model.predict_proba(X)[0]
            print(sign_language_class, sign_language_prob)

            # Grab Ear coordinates
            coords = tuple(np.multiply(np.array((results.pose_landmarks.landmark[mp_holistic.PoseLandmark.LEFT_EAR].x,
                                                 results.pose_landmarks.landmark[mp_holistic.PoseLandmark.LEFT_EAR].y)),
                                       [640, 480]).astype(int))

            # Render prediction
            cv2.rectangle(image,
                          (coords[0], coords[1] + 5),
                          (coords[0] + len(sign_language_class) * 20, coords[1] - 30),
                          (245, 117, 16),
                          -1)
            cv2.putText(image, sign_language_class, coords, cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
            
        except Exception as e:
            print(e)
            pass

        cv2.imshow('Raw Webcam Feed', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
